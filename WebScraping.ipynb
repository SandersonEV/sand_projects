```python
import pandas as pd # work with dataframes
import requests # get the html from the page
from bs4 import BeautifulSoup # Navigate in html code
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
url_html_text = requests.get(url=url).text # get the html file from the url and tranform in text to future calculations
```

```python
soup = BeautifulSoup(url_html_text) # get the html code as text and tranform in a BeautifulSoup Type 
print(type(soup)) # here you can see de type of the data
```

```python
# The <tbody> tag is used to goup the content in an HTML table, so here you'll find the table that you are looking for:

soup_tbody = soup.tbody 
soup_tbody 
```

```python
soup_td = soup_tbody.find_all('tr')
soup_td
```

```python
# When we look to the construction of the HTML table we see that we have two rows and each row represents everything inside of the tag (<tr>...data...<tr>)

construct_dataframe = pd.DataFrame(columns=['Year','Price'])

for row in soup_td: # Here we create a list and for every step (<tr>....<tr>) we will find the data that we need inside of the last <td>...<td>
    col = row.find_all('td') #find all the (<td>...<td>) inside of the (<tr>...<tr>)
    year = col[0].text
    price = col[1].text

    construct_dataframe = construct_dataframe.append({'Year': year,'Price': price}, ignore_index= True)

construct_dataframe
```

```python
# just a alternative way:

construct_dataframe2 = pd.DataFrame(columns=['Year','Price'])

for row in soup.find('tbody').find_all('tr'): # Just a fast way if you still now the exact way and don't need to analyze the html
    col = row.find_all('td') 
    year = col[0].text
    price = col[1].text

    construct_dataframe2 = construct_dataframe2.append({'Year': year,'Price': price}, ignore_index= True)

construct_dataframe2
```

```python
# verify that two ways are the same

construct_dataframe == construct_dataframe2
```

```python
# Remove mising values blank rows and duplicates if it's needed:

construct_dataframe.dropna(inplace= True) # drop the rows if at leat one element of the dataframe is missing
construct_dataframe = construct_dataframe[construct_dataframe['Price'] != ""] # Keep only if the prive is't empty
construct_dataframe['Price'] = construct_dataframe['Price'].str.replace(',|\$',"") # Replace "," and "$" for "" (nothing) to clean the data

construct_dataframe
```

```python
# Change de data type of some columns:
construct_dataframe.dtypes
```

```python
# change data types

construct_dataframe['Year'] = pd.to_numeric(construct_dataframe['Year'])
construct_dataframe['Price'] = pd.to_numeric(construct_dataframe['Price'])
```

```python
# The same as last step (Fast Alternative):

construct_dataframe[['Year','Price']] = construct_dataframe[['Year','Price']].apply(pd.to_numeric)
```
