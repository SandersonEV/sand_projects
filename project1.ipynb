```python
from bs4 import BeautifulSoup as bts # Navegaar através do código html
import requests # "puxar" o html do url fornecido
```

```python
url = "https://pt.wikipedia.org/wiki/Portal:Ci%C3%AAncia"
html_extract = requests.get(url).text # Pega todo o código em html e traz ele e depois tranforma em texto (str)
type(html_extract) # Usando o .text acima, o texto é convertido para string
```

```python
html_extract_bst = bts(html_extract,"html.parser") #tranforma o código de str para um html navegável usando BeautfulSoup
type(html_extract_bst)
```

```python
html_extract_bst.prettify
```

```python
print(html_extract_bst.prettify()) # Mostrar o código html (como o html é muito grande, o codigo que irá aparecer está encurtado)
```

```python
tag_title = html_extract_bst.title #
print(tag_title , "DO TIPO: " , type(tag_title) ) # Percebemos que o tipo da vaiável alocada se refere a um elemento bs4 Tag
```
